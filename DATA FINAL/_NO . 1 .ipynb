```python
# c) What would the values of the standard deviation and IQR be if there was no variability in
#   yearly income for these 4,365 people?                                            
```

```python
no_variability_data = pd.Series([50000] * 4365)
std_dev = no_variability_data.std()
iqr = no_variability_data.quantile(0.75) - no_variability_data.quantile(0.25)
print("Standard Deviation:", std_dev, "IQR:", iqr)
```

```python
# d) What would the values of the mean, median, Q1, Q3, minimum and maximum yearly incomes
#   be if there was no variability?
```

```python
print("Mean:", no_variability_data.mean())
print("Median:", no_variability_data.median())
print("Q1, Q3:", no_variability_data.quantile(0.25), no_variability_data.quantile(0.75))
print("Min, Max:", no_variability_data.min(), no_variability_data.max())
```
